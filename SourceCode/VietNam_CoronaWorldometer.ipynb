```python
import pandas as pd
import os
import warnings
warnings.filterwarnings("ignore")
```

```python
path ='\\Users\\User\\Jupyter Notebook\\DataVisualization\\Lab01'
dirs = os.listdir(path)
listfile = []
for dir in dirs:
    if dir.endswith('.csv'):
        listfile.append(dir)
        
listfile
```

```python
listrow = []
for i in range(len(listfile)):
    tmp = listfile[i].replace('.csv','')
    #print(tmp)
    row = tmp.replace('CoronaVirus','')
    #print(row)
    row = row.replace('2020','')
    listrow.append(row)
    
listrow
```

```python
df = pd.read_csv(os.path.join(path, listfile[0]))

df = df.loc[df['Country']=='Vietnam']
df
```

```python
df_new = df
df_new.insert(2, 'Date', listrow[0])
```

```python
df_new
```

```python
for i in range(1,len(listfile)):
    df_temp = pd.read_csv(os.path.join(path, listfile[i]))
    selected_row = df_temp.loc[df_temp['Country']=='Vietnam']
    selected_row.insert(2, 'Date', listrow[i])
    #print(selected_row)
    df_new = df_new.append(selected_row)
```

```python
df_new
```

```python
df_new = df_new.set_index('Date')
df_new
```

```python
df_new = df_new.drop(['Unnamed: 0','Country','Population','Region'], axis=1)
df_new
```

```python
df_new = df_new.fillna(0)
df_new
```

```python
pathnew = path + '\\Result'
df_new.to_csv(os.path.join(pathnew, 'Vietnam_Corona_Worldometer.csv'))
```
